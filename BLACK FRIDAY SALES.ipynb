# DATA preprocessing

```python
# Drop unnecessary fields

# Convert categorical data into integer using map function (e.g 'Gender' column)

# missing value treatment

# Rename columns

# fill nan values

# map range variables into integers (e.g 'Age' column)
```

```python
# Drop unnecessary fields
df.drop(['User_ID','Product_ID'],axis=1,inplace=True)
```

```python
# Convert categorical data into integer using map function (e.g 'Gender' column)
def gender(i):
    if i=='F':
        return 0
    else:
        return 1

df['Gender']=df['Gender'].map(gender)
```

```python
# missing value treatment
```

```python
# Rename columns
df.rename(columns={'City_Category':'City','Stay_In_Current_City_Years':'Stay_duration'},inplace=True)
```

```python
df['Age']
```

```python
# fill nan values
df.isnull().sum()
```

```python
age=df.groupby('Age')['Product_Category_3']
df['Product_Category_3']=age.transform(lambda x:x.fillna(np.nanmedian(x)))
# age.median()
```

```python
age=df.groupby('Age')['Product_Category_2']
df['Product_Category_2']=age.transform(lambda x:x.fillna(np.nanmedian(x)))
```

```python
df['Age'].unique()
```

```python
# map range variables into integ-ers (e.g 'Age' column)
df['Age']=df['Age'].replace({'0-17':9,'18-25':21,'26-35':31,'36-45':41,'46-50':48,'51-55':52,'55+':70})
```
